**Comparative Text Classification using Statistical and Embedding-Based Models - Sentiment Analysis of Movie Reviews**

**Introduction to the Domain-Specific Area-Sentiment Analysis of Movie Reviews**

Sentiment analysis is an important area within Natural Language Processing that aims to detect and categorize opinions in text as positive, negative, or neutral. When it is applied to move reviews, it helps to understand how audiences feel about films. This is beneficial as it offers valuable insights for movie studios, streaming services, and recommendation engines. Online platforms are now producing a huge number of user reviews, so it is no longer practical to analyze them by hand. Because of this, automated sentimental analysis has become essential. Public move review datasets, such as those available on Kaggle, provide real and labeled data that help to build and test sentiment analysis models.

However there are also several challenges involved with sentiment analysis of movie reviews. Many reviews contain informal language, slang, sarcasm, and idiomatic expressions that are difficult for computational models to interpret correctly. Sentiment is also highly context-dependent; a single review may contain both praise and criticism, or the tone may change from one sentence to another. Problems like negation, movie-related terms, and big differences in review length makes accurate classification harder. Because of this, many traditional methods have trouble working well across different writing styles and opinions.

To handle these challenges, researchers have tried different approaches over the years, leading to the steady development of sentiment analysis over the past two decades.

In the early 2000s, researchers mainly used rule-based methods and sentiment word lists to classify opinions. Later, machine learning models such as Naïve Bayes and Support Vector Machines proved more effective by learning patterns from data. However, these models treated text as unordered words and struggled to understand context.

Between 2010 and 2015, the release of large labeled datasets greatly improved research in this field. The IMDB movie review dataset allowed more reliable training and evaluation of sentiment models. During this period, the Stanford Sentiment Treebank introduced phrase-level sentiment labels, showing that sentence structure and word combinations play an important role in understanding sentiment.

From 2014 to 2018, deep learning models brought major improvements to sentiment analysis. Convolutional Neural Networks captured local patterns in text, while Recurrent Neural Networks and LSTMs modeled word order and long-term dependencies. These models handled complex language features such as negation more effectively than earlier methods.

Since 2018, transformer-based models have become the dominant approach in sentiment analysis. Models like BERT learn rich contextual representations from large text corpora and can be fine-tuned for specific domains, leading to higher accuracy and better generalization. At the same time, research has increasingly focused on domain-specific challenges. Movie reviews often contain mixed opinions about different aspects of a film, which led to the development of aspect-based sentiment analysis. Social media sentiment analysis also gained importance, requiring models to handle informal language, slang, and emojis.

**Section 4: Evaluation Methodology**

To measure how well the sentiment analysis models perform on the IMDB 50K Movie Reviews dataset, several evaluation metrics are used to give a clear and complete picture of model quality. Relying on more than one metric is important because a single measure cannot fully describe how well a model handles different types of errors.

Accuracy shows how often the model predicts the correct sentiment overall. It is calculated by dividing the number of correct predictions by the total number of reviews. Although accuracy can sometimes be misleading when one class is much larger than the other, the IMDB dataset is evenly split between positive and negative reviews, so accuracy is a useful starting point in this case.

Precision looks at how accurate the model’s positive predictions are. It tells us how many reviews labeled as positive are actually positive. High precision is important when false positives could give a wrong impression of audience opinions.

Recall measures how well the model finds all the truly positive reviews. It shows the percentage of actual positive reviews that the model correctly identifies. This is important when missing positive feedback could lead to an incomplete understanding of audience response.

The F1-score combines precision and recall into a single value, balancing both types of errors. It is especially helpful when comparing models that may perform well on one metric but poorly on another.

To ensure a fair comparison, both the traditional statistical model and the deep learning model are trained and tested using the same data split. Their performance is then compared using the test results. This makes it easier to understand the trade-offs between simple, interpretable models and more complex ones with higher predictive power.

Cross-validation may also be used during training to reduce overfitting and help the models generalize better to new data. Since the dataset is balanced, no extra steps like resampling or class weighting are needed. Overall, this evaluation approach allows for a fair and reliable comparison between traditional and deep learning methods for sentiment analysis.

```python
# ============================================================================
# Text Cleaning and Preprocessing
# ============================================================================

def clean_review_text(text):
    """
    Clean and normalize a single movie review.

    The goal of this function is to remove noise from the raw text while
    preserving words that are useful for sentiment analysis.

    Processing steps:
      1. Remove HTML tags
      2. Convert text to lowercase
      3. Remove punctuation and numbers
      4. Tokenize text into words
      5. Remove stopwords and very short words
      6. Lemmatize words to their base form

    Returns:
        A cleaned string ready for feature extraction.
    """

    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)

    # Convert text to lowercase
    text = text.lower()

    # Remove punctuation and numbers, keep only letters
    text = re.sub(r'[^a-z\s]', '', text)

    # Tokenize text into individual words
    words = word_tokenize(text)

    # Remove stopwords and short words
    stop_words = set(stopwords.words('english'))
    words = [w for w in words if w not in stop_words and len(w) > 2]

    # Lemmatize words to their base form
    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(w) for w in words]

    # Reconstruct cleaned text
    return ' '.join(words)


def clean_all_reviews(df):
    """
    Apply text cleaning to all reviews in the dataset.

    This function:
      • Cleans every review using `clean_review_text`
      • Converts sentiment labels to numeric format
      • Reports processing time
      • Displays a before-and-after example
      • Outputs basic vocabulary statistics
    """

    print("\n" + "=" * 70)
    print("CLEANING ALL REVIEWS")
    print("=" * 70)

    start = time.time()
    print("\nProcessing reviews...")

    # Apply text cleaning to all reviews
    df['cleaned_review'] = df['review'].apply(clean_review_text)

    # Convert sentiment labels to binary values
    df['label'] = df['sentiment'].map({'negative': 0, 'positive': 1})

    elapsed_time = time.time() - start
    print(f"Cleaning completed in {elapsed_time:.1f} seconds")

    # Display a sample before-and-after comparison
    sample_idx = 100
    print("\nExample review transformation:")
    print("\nBefore cleaning:")
    print(df['review'].iloc[sample_idx][:250] + "...")
    print("\nAfter cleaning:")
    print(df['cleaned_review'].iloc[sample_idx][:250] + "...")

    # Compute vocabulary statistics
    all_words = ' '.join(df['cleaned_review']).split()
    unique_words = set(all_words)

    print("\nVocabulary statistics:")
    print(f"  Total words: {len(all_words):,}")
    print(f"  Unique words: {len(unique_words):,}")
    print(f"  Average words per review: {len(all_words) / len(df):.1f}")

    return df


# Execute text cleaning on the dataset
df = clean_all_reviews(df)
```

**Part 3: Cleaning and Preprocessing the Text**

Text preprocessing converts raw, noisy movie reviews into clean, standardized text that machine learning models can understand. The clean_review_text function handles a single review through several steps:

1. Remove HTML tags using a regular expression. Many reviews contain tags, which do not carry sentiment and could confuse the model.

2. Convert text to lowercase so that words like “Good”, “GOOD”, and “good” are treated the same. This reduces the vocabulary size and improves feature consistency.

3. Remove punctuation and numbers, keeping only lowercase letters and spaces. This focuses the text on meaningful content words.

4. Tokenization splits the text into individual words using NLTK’s word_tokenize, which handles contractions and punctuation properly (e.g., “don’t” → “do” + “n’t”).

5. Remove stopwords—common words like “the”, “is”, and “at” that carry little sentiment meaning. Removing them reduces dimensionality and emphasizes important words.

6. Lemmatization reduces words to their base forms using WordNet. For example, “running”, “runs”, and “ran” all become “run”, and “better” or “best” become “good”. This groups semantically similar words, reduces vocabulary size, and helps the model generalize.

7. Rejoin tokens into a single cleaned string for each review.

The clean_all_reviews function applies this process to every review in the dataset, creating a new column called cleaned_review. Before-and-after examples show how the text is transformed, providing transparency and verifying that the pipeline works as intended.

Finally, vocabulary statistics are computed by counting total and unique words across all cleaned reviews. This ensures preprocessing is not too aggressive while reducing the vocabulary to a manageable size, which is important for vectorizers and downstream modeling. A typical result might show about 30,000 unique words after cleaning, guiding decisions about feature selection.

```python
# ============================================================================
# Splitting the Data
# ============================================================================

def split_data(df, test_size=0.2):
    """
    Split the dataset into training and testing sets.

    The model is trained on one part of the data and tested on reviews it has
    not seen before. This shows how well it performs on new, unseen data.

    Eighty percent of the data is used for training and twenty percent
    for testing, which is a common split.
    """
    print("\n" + "="*70)
    print("SPLITTING INTO TRAINING AND TEST SETS")
    print("="*70)

    # X is the input (the review text)
    # y is the output (positive or negative)
    X = df['cleaned_review']
    y = df['label']

    # Split the data, making sure both sets have equal positive/negative balance
    X_train, X_test, y_train, y_test = train_test_split(
        X, y,
        test_size=test_size,
        random_state=42,
        stratify=y
    )

    print(f"\nData split complete:")
    print(f"  Training set: {len(X_train):,} reviews ({(1-test_size)*100:.0f}%)")
    print(f"  Test set: {len(X_test):,} reviews ({test_size*100:.0f}%)")
    print(f"\n  Positive reviews in training: {y_train.sum():,} ({y_train.mean()*100:.1f}%)")
    print(f"  Positive reviews in test: {y_test.sum():,} ({y_test.mean()*100:.1f}%)")

    return X_train, X_test, y_train, y_test


# EXECUTE: Split the data
X_train, X_test, y_train, y_test = split_data(df)
```

**Part 4: Splitting the Data**

Data splitting separates the dataset into training and testing portions. The training set teaches models to recognize patterns, while the test set evaluates performance on unseen reviews, simulating real-world usage.

The code uses scikit-learn’s train_test_split with key parameters for fairness and reproducibility:

* test_size=0.2: 20% of the data is reserved for testing, leaving 80% for training. This 80-20 split is standard, giving the models enough data to learn and the test set enough examples (10,000 reviews) for reliable evaluation.

* random_state=42: fixes the random seed, ensuring the split is the same every time the code runs. This makes model comparisons fair and reproducible.

* stratify=y: maintains the same class distribution in both sets. Each set has 50% positive and 50% negative reviews, preventing small imbalances that could affect training or evaluation.

Before splitting, the code extracts features (X) and labels (y):

* Features (X) are the cleaned review texts that serve as inputs to the models.

* Labels (y) are converted to integers: 1 for positive and 0 for negative sentiment. Numeric labels are required for all machine learning models.

After splitting, the function prints the size of each set and confirms class balance. This validation ensures the split worked correctly. The function returns four arrays: X_train, X_test, y_train, and y_test, which are used for training and evaluating all models.

This step ensures a fair and consistent setup for comparing Logistic Regression, Naive Bayes, and LSTM models, making sure differences in performance reflect the models themselves, not variations in the data.

```python
# ============================================================================
# First Model - Simple Logistic Regression
# ============================================================================

def train_baseline_model(X_train, X_test, y_train, y_test):
    """
    Build a simple baseline model using Logistic Regression.

    This is a simple method using a "bag of words" approach, which
    just counts how many times each word appears, ignoring word order.

    It's simple but often surprisingly effective.
    """
    print("\n" + "="*70)
    print("MODEL 1: BASELINE (LOGISTIC REGRESSION WITH BAG OF WORDS)")
    print("="*70)

    start = time.time()

    # Create a bag of words representation
    print("\nConverting text to numbers using Bag of Words...")
    vectorizer = CountVectorizer(max_features=3000)

    # Fit on training data and transform both sets
    X_train_bow = vectorizer.fit_transform(X_train)
    X_test_bow = vectorizer.transform(X_test)

    print(f"Features created:")
    print(f"  Number of unique words used: {len(vectorizer.get_feature_names_out()):,}")
    print(f"  Training data shape: {X_train_bow.shape}")

    # Train the logistic regression model
    print(f"\nTraining the model...")
    model = LogisticRegression(max_iter=1000, random_state=42)
    model.fit(X_train_bow, y_train)

    # Make predictions on the test set
    y_pred = model.predict(X_test_bow)

    train_time = time.time() - start

    # See the results
    acc = accuracy_score(y_test, y_pred)
    prec, rec, f1, _ = precision_recall_fscore_support(y_test, y_pred, average='binary')

    print(f"\nTraining complete in {train_time:.1f} seconds")
    print(f"\nResults:")
    print(f"  Accuracy: {acc:.4f} ({acc*100:.2f}%)")
    print(f"  Precision: {prec:.4f}")
    print(f"  Recall: {rec:.4f}")
    print(f"  F1-Score: {f1:.4f}")

    print(f"\nDetailed breakdown:")
    print(classification_report(y_test, y_pred,
                               target_names=['Negative', 'Positive']))

    cm = confusion_matrix(y_test, y_pred)

    return {
        'model': model,
        'vectorizer': vectorizer,
        'predictions': y_pred,
        'accuracy': acc,
        'precision': prec,
        'recall': rec,
        'f1': f1,
        'training_time': train_time,
        'confusion_matrix': cm
    }


# EXECUTE: Train baseline model
baseline_results = train_baseline_model(X_train, X_test, y_train, y_test)
```

**Part 5: Baseline Model – Logistic Regression**

This section implements Logistic Regression as a baseline model. It is fast, interpretable, and surprisingly effective for text classification. The model uses a Bag-of-Words approach, converting text into feature vectors based on word counts while ignoring word order. Although this method treats “not good” the same as “good not,” it works well because sentiment is often expressed through clear keywords.

The CountVectorizer converts text into numeric features with two key parameters:

* max_features=10000 limits the vocabulary to the 10,000 most frequent words, reducing noise and computational cost.

* ngram_range=(1,1) uses unigrams (single words) only, keeping the baseline simple.

Vectorization occurs in two steps:

1. fit_transform on training data builds the vocabulary and converts reviews into a sparse matrix where rows are reviews and columns are word counts.

2. transform on test data uses the same vocabulary, ensuring the test set remains unseen.

The Logistic Regression model is configured with max_iter=1000 for convergence and random_state for reproducibility. During training, the model learns weights for each word to predict sentiment. Predictions compute a weighted sum of word counts, apply a sigmoid function, and classify reviews as positive (≥0.5) or negative (<0.5).

Model evaluation uses multiple metrics:

* Accuracy measures overall correctness.

* Precision shows the accuracy of positive predictions.

* Recall measures how many true positives were identified.

* F1-score balances precision and recall.

* Classification report provides per-class metrics.

* Confusion matrix shows true positives, true negatives, false positives, and false negatives, giving insight into error patterns.

Results are stored for comparison with more advanced models. Typical baseline performance achieves around 88% accuracy, providing a strong reference point for subsequent models like Naive Bayes and LSTM.

```python
# ============================================================================
# Improved Statistical Model - Naive Bayes with TF-IDF
# ============================================================================

def train_statistical_model(X_train, X_test, y_train, y_test):
    """
    An improved statistical model is built by combining TF-IDF with Naive Bayes.
     TF-IDF is more effective than simple word counting because it assigns
     higher importance to words that carry meaningful information in a review
     but are not common across all reviews, such as “excellent,” while reducing
      the influence of frequent but less informative words like “movie.”
      Naive Bayes is a probabilistic model that performs well with text data,
      making it a suitable choice for this approach.
    """
    print("\n" + "="*70)
    print("MODEL 2: NAIVE BAYES WITH TF-IDF")
    print("="*70)

    start = time.time()

    # Create TF-IDF features
    print("\nConverting text using TF-IDF (smarter word weighting)...")
    tfidf = TfidfVectorizer(
        max_features=5000,      # Use top 5000 words
        ngram_range=(1, 2),     # Look at single words and pairs
        min_df=5,               # Word must appear in at least 5 reviews
        max_df=0.7,             # Ignore words in more than 70% of reviews
        sublinear_tf=True       # Use logarithmic term frequency scaling
    )

    X_train_tfidf = tfidf.fit_transform(X_train)
    X_test_tfidf = tfidf.transform(X_test)

    print(f"TF-IDF features created:")
    print(f"  Vocabulary size: {len(tfidf.get_feature_names_out()):,}")
    print(f"  Training matrix shape: {X_train_tfidf.shape}")

    # Train Naive Bayes
    print(f"\nTraining Naive Bayes model...")
    model = MultinomialNB(alpha=0.1)
    model.fit(X_train_tfidf, y_train)

    # Make predictions
    y_pred = model.predict(X_test_tfidf)

    train_time = time.time() - start

    # Evaluate performance
    acc = accuracy_score(y_test, y_pred)
    prec, rec, f1, _ = precision_recall_fscore_support(y_test, y_pred, average='binary')

    print(f"\nTraining complete in {train_time:.1f} seconds")
    print(f"\nResults:")
    print(f"  Accuracy: {acc:.4f} ({acc*100:.2f}%)")
    print(f"  Precision: {prec:.4f}")
    print(f"  Recall: {rec:.4f}")
    print(f"  F1-Score: {f1:.4f}")

    print(f"\nDetailed breakdown:")
    print(classification_report(y_test, y_pred,
                               target_names=['Negative', 'Positive']))

    # Find the most important words for each sentiment
    feature_names = tfidf.get_feature_names_out()

    # Top words for positive reviews
    top_pos_idx = model.feature_log_prob_[1].argsort()[-10:][::-1]
    top_positive = [feature_names[i] for i in top_pos_idx]

    # Top words for negative reviews
    top_neg_idx = model.feature_log_prob_[0].argsort()[-10:][::-1]
    top_negative = [feature_names[i] for i in top_neg_idx]

    print(f"\nMost important words for POSITIVE sentiment:")
    print(f"  {', '.join(top_positive)}")

    print(f"\nMost important words for NEGATIVE sentiment:")
    print(f"  {', '.join(top_negative)}")

    # Visualize results with confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=True,
                xticklabels=['Negative', 'Positive'],
                yticklabels=['Negative', 'Positive'])
    plt.title('Confusion Matrix - Naive Bayes Model', fontweight='bold', fontsize=14)
    plt.ylabel('Actual Sentiment', fontsize=12)
    plt.xlabel('Predicted Sentiment', fontsize=12)
    plt.tight_layout()
    plt.savefig('nb_confusion_matrix.png', dpi=300, bbox_inches='tight')
    plt.show()

    return {
        'model': model,
        'vectorizer': tfidf,
        'predictions': y_pred,
        'accuracy': acc,
        'precision': prec,
        'recall': rec,
        'f1': f1,
        'training_time': train_time,
        'confusion_matrix': cm
    }

# EXECUTE: Train statistical model
statistical_results = train_statistical_model(X_train, X_test, y_train, y_test)
```

**Part 6: Statistical Model – Naive Bayes with TF-IDF**

This section implements an enhanced statistical model using TF-IDF features with a Multinomial Naive Bayes classifier. TF-IDF (Term Frequency-Inverse Document Frequency) improves on simple word counts by weighting words based on how informative they are. Words that appear often in a review but rarely across all reviews receive higher weights. For example, “excellent” might get a high score, while common words like “movie” get low scores.

The TfidfVectorizer includes both unigrams and bigrams (ngram_range=(1,2)) to capture context. Bigrams preserve sentiment nuances that unigrams miss: phrases like “not good” or “very good” carry meaning beyond individual words. Common bigrams in reviews include “waste time,” “highly recommend,” “plot holes,” and “character development.” To manage feature size, max_features=10000 limits the model to the most informative n-grams.

Multinomial Naive Bayes is chosen because it works well for text classification. It assumes words are independent given the class (“naive”), which is not strictly true but still effective. The multinomial variant models word counts, which works naturally with TF-IDF. Training is extremely fast because it only requires counting word occurrences and computing probabilities—no iterative optimization is needed.

Evaluation includes feature importance, showing which n-grams indicate positive or negative sentiment. Positive indicators include “excellent,” “brilliant,” and “highly recommend,” while negative indicators include “terrible,” “boring,” and “waste time.” This interpretability helps validate that the model learns meaningful patterns rather than random correlations.

The confusion matrix is visualized as a heatmap, highlighting error patterns. It makes false positives and false negatives easy to see and shows how well the model performs on each class. Comparing with the baseline Logistic Regression reveals how TF-IDF and bigrams reduce errors.

This improved model typically achieves 89–90% accuracy, a modest but meaningful improvement over the baseline’s 88%. It demonstrates that careful feature engineering—using TF-IDF and bigrams—can enhance performance without increasing model complexity.

```python
# ============================================================================
# Deep Learning Model - Bidirectional LSTM
# ============================================================================

def train_deep_learning_model(X_train, X_test, y_train, y_test):
    """
    A neural network is built using a Bidirectional LSTM. Unlike simpler models,
    this approach understands word order and context by reading the text both
    forwards and backwards, allowing it to recognize patterns such as
    “not good,” which is negative despite the presence of a usually positive
    word. This model is more powerful but requires significantly more time
    to train.
    """
    print("\n" + "="*70)
    print("MODEL 3: DEEP LEARNING (BIDIRECTIONAL LSTM)")
    print("="*70)

    # Settings for the neural network
    MAX_WORDS = 10000  # Vocabulary size
    MAX_LEN = 200      # Maximum review length (in words)
    EMBEDDING_DIM = 128  # Size of word vectors
    BATCH_SIZE = 64    # Reviews to process at once
    EPOCHS = 10        # Training iterations

    # Prepare the text for the neural network
    print("\nPreparing text for neural network...")
    tokenizer = Tokenizer(num_words=MAX_WORDS, oov_token='<OOV>')
    tokenizer.fit_on_texts(X_train)

    # Convert text to sequences of numbers
    X_train_seq = tokenizer.texts_to_sequences(X_train)
    X_test_seq = tokenizer.texts_to_sequences(X_test)

    # Make all sequences the same length
    X_train_pad = pad_sequences(X_train_seq, maxlen=MAX_LEN, padding='post')
    X_test_pad = pad_sequences(X_test_seq, maxlen=MAX_LEN, padding='post')

    print(f"Text converted to sequences")
    print(f"  Training data shape: {X_train_pad.shape}")
    print(f"  Each review represented as {MAX_LEN} numbers")

    # Build the neural network architecture
    print("\nBuilding the neural network...")
    model = Sequential([
        # First layer: learns word meanings
        Embedding(MAX_WORDS, EMBEDDING_DIM, input_length=MAX_LEN),

        # Main layer: LSTM that reads forwards and backwards
        Bidirectional(LSTM(64)),

        # Dropout to prevent overfitting
        Dropout(0.5),

        # Output layer: positive or negative
        Dense(1, activation='sigmoid')
    ])

    # Set up how the model learns
    model.compile(
        optimizer='adam',
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    print("\nModel architecture:")
    model.summary()

    # Set up early stopping if no improvement is seen
    early_stop = EarlyStopping(
        monitor='val_loss',
        patience=3,
        restore_best_weights=True
    )

    # Train the model
    print("\nTraining the neural network (this takes a while)...")
    start = time.time()

    history = model.fit(
        X_train_pad, y_train,
        batch_size=BATCH_SIZE,
        epochs=EPOCHS,
        validation_split=0.2,
        callbacks=[early_stop],
        verbose=1
    )

    train_time = time.time() - start

    # Make predictions
    y_pred = (model.predict(X_test_pad) > 0.5).astype(int).flatten()

    # Evaluate
    acc = accuracy_score(y_test, y_pred)
    prec, rec, f1, _ = precision_recall_fscore_support(y_test, y_pred, average='binary')

    print(f"\n\nTraining complete in {train_time:.1f} seconds ({train_time/60:.1f} minutes)")
    print(f"\nResults:")
    print(f"  Accuracy: {acc:.4f} ({acc*100:.2f}%)")
    print(f"  Precision: {prec:.4f}")
    print(f"  Recall: {rec:.4f}")
    print(f"  F1-Score: {f1:.4f}")

    print(f"\nDetailed breakdown:")
    print(classification_report(y_test, y_pred,
                               target_names=['Negative', 'Positive']))

    # Visualize confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Greens', cbar=True,
                xticklabels=['Negative', 'Positive'],
                yticklabels=['Negative', 'Positive'])
    plt.title('Confusion Matrix - LSTM Model', fontweight='bold', fontsize=14)
    plt.ylabel('Actual Sentiment', fontsize=12)
    plt.xlabel('Predicted Sentiment', fontsize=12)
    plt.tight_layout()
    plt.savefig('lstm_confusion_matrix.png', dpi=300, bbox_inches='tight')
    plt.show()

    # Visualize training progress
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

    # Accuracy over time
    ax1.plot(history.history['accuracy'], label='Training', marker='o', linewidth=2)
    ax1.plot(history.history['val_accuracy'], label='Validation', marker='s', linewidth=2)
    ax1.set_title('Model Accuracy During Training', fontweight='bold', fontsize=12)
    ax1.set_xlabel('Epoch', fontsize=11)
    ax1.set_ylabel('Accuracy', fontsize=11)
    ax1.legend(loc='lower right')
    ax1.grid(alpha=0.3)

    # Loss over time
    ax2.plot(history.history['loss'], label='Training', marker='o', linewidth=2)
    ax2.plot(history.history['val_loss'], label='Validation', marker='s', linewidth=2)
    ax2.set_title('Model Loss During Training', fontweight='bold', fontsize=12)
    ax2.set_xlabel('Epoch', fontsize=11)
    ax2.set_ylabel('Loss', fontsize=11)
    ax2.legend(loc='upper right')
    ax2.grid(alpha=0.3)

    plt.tight_layout()
    plt.savefig('lstm_training_history.png', dpi=300, bbox_inches='tight')
    plt.show()

    return {
        'model': model,
        'tokenizer': tokenizer,
        'accuracy': acc,
        'precision': prec,
        'recall': rec,
        'f1': f1,
        'training_time': train_time,
        'predictions': y_pred,
        'confusion_matrix': cm,
        'history': history.history
    }


# EXECUTE: Train deep learning model
lstm_results = train_deep_learning_model(X_train, X_test, y_train, y_test)
```

**Part 7: Deep Learning Model – Bidirectional LSTM**

This section implements a deep learning model using a bidirectional LSTM to capture sequential patterns in text. Unlike bag-of-words models, LSTMs process reviews as sequences, understanding that phrases like “not good” differ from “good not.” LSTMs handle long-range dependencies with gates that control what information to remember or forget, avoiding the vanishing gradient problem that affects standard RNNs.

The bidirectional LSTM reads sequences forwards and backwards, incorporating both past and future context. For example, in “The movie started slow but ended brilliantly,” it sees the words after “slow,” allowing better understanding of negation, contrast, or subtle sentiment.

Text is converted to integer sequences using a Keras Tokenizer limited to the 10,000 most frequent words. An oov_token handles unknown words in test data. Sequences are padded or truncated to 200 tokens based on review length analysis, ensuring consistent input size without wasting computation.

The network has five layers:

1. Embedding layer: converts word indices to 128-dimensional dense vectors, capturing semantic similarity.

2. Bidirectional LSTM: 64 units per direction, producing a 128-dimensional review representation.

3. Dropout (30%): prevents overfitting.

4. Dense layer with ReLU: learns high-level features.

5. Output layer with sigmoid: produces a probability for positive sentiment.

The model trains for 5 epochs with batch size 64 and a 20% validation split to monitor overfitting. Adam optimizer adjusts learning rates automatically, and binary cross-entropy is used as the loss function. Training is much longer than statistical models—5–10 minutes on GPU or 30–60 minutes on CPU.

Typical performance reaches 90–91% accuracy, slightly higher than statistical models but at much higher computational cost. This modest gain suggests that for IMDB reviews, most sentiment is captured by word presence, so deep contextual understanding provides only limited improvement.

```python
# ============================================================================
# CELL 8: Comparing All Models
# ============================================================================

def compare_all_models(baseline_results, nb_results, lstm_results, y_test):
    """
    All three models are now compared side by side to evaluate their
    performance. This comparison makes it possible to see which model
    achieves the highest accuracy, which one trains the fastest, and how
    accuracy and training time trade off against each other. It also
    helps determine the situations in which each approach is most
    appropriate to use.
    """
    print("\n" + "="*70)
    print("COMPARING ALL THREE MODELS")
    print("="*70)

    # Organize results into a table
    results = {
        'Model': [
            'Baseline (Logistic Regression)',
            'Naive Bayes (TF-IDF)',
            'LSTM (Deep Learning)'
        ],
        'Accuracy': [
            baseline_results['accuracy'],
            nb_results['accuracy'],
            lstm_results['accuracy']
        ],
        'Precision': [
            baseline_results['precision'],
            nb_results['precision'],
            lstm_results['precision']
        ],
        'Recall': [
            baseline_results['recall'],
            nb_results['recall'],
            lstm_results['recall']
        ],
        'F1-Score': [
            baseline_results['f1'],
            nb_results['f1'],
            lstm_results['f1']
        ],
        'Training Time (seconds)': [
            baseline_results['training_time'],
            nb_results['training_time'],
            lstm_results['training_time']
        ]
    }

    results_df = pd.DataFrame(results)

    print("\nPERFORMANCE SUMMARY:")
    print(results_df.to_string(index=False))

    # Create comparison visualizations
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

    # Chart 1: Compare all metrics
    metrics = results_df.set_index('Model')[['Accuracy', 'Precision', 'Recall', 'F1-Score']]
    metrics.plot(kind='bar', ax=ax1, width=0.8, rot=15)
    ax1.set_title('Model Performance Comparison', fontweight='bold', fontsize=14)
    ax1.set_ylabel('Score', fontsize=12)
    ax1.set_xlabel('Model', fontsize=12)
    ax1.set_ylim([0.75, 1.0])
    ax1.legend(loc='lower right', fontsize=10)
    ax1.grid(axis='y', alpha=0.3)
    ax1.axhline(y=0.9, color='red', linestyle='--', alpha=0.5)

    # Chart 2: Compare training times
    models = results_df['Model']
    times = results_df['Training Time (seconds)']
    colors = ['#ff6b6b', '#4ecdc4', '#95e1d3']

    ax2.bar(models, times, color=colors, edgecolor='black', alpha=0.8)
    ax2.set_title('Training Time Comparison', fontweight='bold', fontsize=14)
    ax2.set_ylabel('Time (seconds, log scale)', fontsize=12)
    ax2.set_xlabel('Model', fontsize=12)
    ax2.set_yscale('log')  # Log scale because LSTM takes much longer
    ax2.set_xticklabels(models, rotation=15, ha='right')
    ax2.grid(axis='y', alpha=0.3)

    # Add time labels on bars
    for i, (model, t) in enumerate(zip(models, times)):
        label = f'{t:.1f}s' if t < 60 else f'{t/60:.1f}m'
        ax2.text(i, t, label, ha='center', va='bottom', fontsize=10, fontweight='bold')

    plt.tight_layout()
    plt.savefig('model_comparison.png', dpi=300, bbox_inches='tight')
    plt.show()

    # Show all confusion matrices side by side
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))

    cms = [
        baseline_results['confusion_matrix'],
        nb_results['confusion_matrix'],
        lstm_results['confusion_matrix']
    ]

    titles = ['Baseline', 'Naive Bayes', 'LSTM']
    cmaps = ['Reds', 'Blues', 'Greens']

    for idx, (cm, title, cmap) in enumerate(zip(cms, titles, cmaps)):
        sns.heatmap(cm, annot=True, fmt='d', cmap=cmap, ax=axes[idx],
                    xticklabels=['Negative', 'Positive'],
                    yticklabels=['Negative', 'Positive'],
                    cbar=True)
        acc = results_df.iloc[idx]['Accuracy']
        axes[idx].set_title(f'{title}\nAccuracy: {acc:.3f}',
                          fontweight='bold', fontsize=12)
        axes[idx].set_ylabel('Actual' if idx == 0 else '', fontsize=11)
        axes[idx].set_xlabel('Predicted', fontsize=11)

    plt.tight_layout()
    plt.savefig('all_confusion_matrices.png', dpi=300, bbox_inches='tight')
    plt.show()

    # Analyze improvements
    print("\nPERFORMANCE IMPROVEMENTS:")
    base_acc = baseline_results['accuracy']
    nb_improvement = (nb_results['accuracy'] - base_acc) / base_acc * 100
    lstm_improvement = (lstm_results['accuracy'] - base_acc) / base_acc * 100

    print(f"  Naive Bayes vs Baseline: {nb_improvement:+.2f}%")
    print(f"  LSTM vs Baseline: {lstm_improvement:+.2f}%")

    nb_vs_lstm = (lstm_results['accuracy'] - nb_results['accuracy']) * 100
    print(f"  LSTM vs Naive Bayes: {nb_vs_lstm:+.2f} percentage points")

    # Error analysis
    base_errors = len(y_test) * (1 - base_acc)
    nb_errors = len(y_test) * (1 - nb_results['accuracy'])
    lstm_errors = len(y_test) * (1 - lstm_results['accuracy'])

    print(f"\nERROR COUNTS:")
    print(f"  Baseline: {base_errors:.0f} wrong predictions")
    print(f"  Naive Bayes: {nb_errors:.0f} wrong ({base_errors - nb_errors:+.0f})")
    print(f"  LSTM: {lstm_errors:.0f} wrong ({base_errors - lstm_errors:+.0f})")

    return results_df


# EXECUTE: Compare all models
results_df = compare_all_models(baseline_results, statistical_results,
                                 lstm_results, y_test)
```

```python
# ============================================================================
# Detailed Analysis and Discussion
# ============================================================================

def detailed_analysis(baseline_results, statistical_results, lstm_results,
                      results_df, y_test):
    """
    This function provides a detailed analysis of model performance and offers
    guidance on when each type of model is most appropriate.
    """
    print("\n" + "="*70)
    print("DETAILED PERFORMANCE ANALYSIS")
    print("="*70)

    # Display summary of results
    print("\nSUMMARY OF RESULTS:")
    print(results_df.to_string(index=False))

    # Calculate and display performance range
    accuracies = [baseline_results['accuracy'],
                  statistical_results['accuracy'],
                  lstm_results['accuracy']]

    print("\n" + "="*70)
    print("INTERPRETATION OF RESULTS")
    print("="*70)

    print(f"\n1. PERFORMANCE COMPARISON:")
    print(f"   The models achieve accuracies between {min(accuracies):.4f} and {max(accuracies):.4f}")
    print(f"   The maximum difference is {(max(accuracies) - min(accuracies))*100:.2f} percentage points")
    print(f"   This is because the IMDB dataset contains clear sentiment indicators such as")
    print(f"   'excellent', 'terrible', 'amazing', and 'awful', which simple models can detect.")

    print(f"\n2. COMPUTATIONAL EFFICIENCY:")
    times = [baseline_results['training_time'],
             statistical_results['training_time'],
             lstm_results['training_time']]

    print(f"   Baseline model: {times[0]:.1f}s - very fast")
    print(f"   Naive Bayes: {times[1]:.1f}s - still efficient")
    print(f"   LSTM: {times[2]:.1f}s ({times[2]/60:.1f} minutes) - significantly slower")
    print(f"   The LSTM requires {times[2]/times[0]:.0f} times more training time than the baseline")
    print(f"   while providing only {(accuracies[2]-accuracies[0])*100:.2f} percentage points improvement in accuracy")

    print(f"\n3. ERROR REDUCTION:")
    errors = [10000 * (1 - acc) for acc in accuracies]
    print(f"   Baseline errors: {errors[0]:.0f}")
    print(f"   Naive Bayes errors: {errors[1]:.0f} (change: {errors[1]-errors[0]:+.0f})")
    print(f"   LSTM errors: {errors[2]:.0f} (change: {errors[2]-errors[0]:+.0f})")

    # Confusion matrix analysis
    print("\n" + "="*70)
    print("DETAILED ERROR ANALYSIS")
    print("="*70)

    models = ['Baseline', 'Naive Bayes', 'LSTM']
    cms = [baseline_results['confusion_matrix'],
           statistical_results['confusion_matrix'],
           lstm_results['confusion_matrix']]

    for name, cm in zip(models, cms):
        tn, fp, fn, tp = cm.ravel()
        sensitivity = tp / (tp + fn)
        specificity = tn / (tn + fp)

        print(f"\n{name}:")
        print(f"   True positives: {tp:,}")
        print(f"   True negatives: {tn:,}")
        print(f"   False negatives: {fn:,}")
        print(f"   False positives: {fp:,}")
        print(f"   Sensitivity: {sensitivity:.4f}")
        print(f"   Specificity: {specificity:.4f}")
        print(f"   Balance between sensitivity and specificity: {abs(sensitivity - specificity):.4f}")

    # Model selection guidance
    print("\n" + "="*70)
    print("RECOMMENDATIONS FOR MODEL SELECTION")
    print("="*70)

    print("\nUse statistical models (e.g., Logistic Regression, Naive Bayes) when:")
    print("   • Fast training and predictions are required")
    print("   • Interpretability is important")
    print("   • Computing resources are limited")
    print("   • Sentiment is explicit and straightforward")
    print("   • Deployment in real-time or on edge devices is needed")

    print("\nUse deep learning models (e.g., LSTM) when:")
    print("   • Access to high-performance hardware is available")
    print("   • Training time is not a concern")
    print("   • Sentiment is subtle or context-dependent")
    print("   • Large datasets are available")
    print("   • Small improvements in accuracy justify higher cost")

    print("\n" + "="*70)
    print("KEY INSIGHTS")
    print("="*70)

    print("\nInsight 1: Simpler models are often sufficient")
    print("   Logistic regression achieves nearly the same accuracy as LSTM")
    print("   while training significantly faster.")

    print("\nInsight 2: Dataset characteristics affect model choice")
    print("   IMDB reviews contain explicit sentiment words, making simple models effective.")

    print("\nInsight 3: Context awareness has limited benefit for clear sentiment")
    print("   LSTM captures word order and context, but for basic positive/negative sentiment")
    print("   detection, the improvement is marginal.")

    print("\nInsight 4: Consider practical constraints")
    print("   In most real-world applications, Naive Bayes is preferred due to its balance of")
    print("   accuracy, speed, and interpretability unless LSTM is specifically required.")

    # Visualization of performance trade-offs
    fig, ax = plt.subplots(1, 1, figsize=(10, 6))
    model_names = ['Baseline', 'Naive Bayes', 'LSTM']
    colors = ['#ff6b6b', '#4ecdc4', '#95e1d3']

    times_log = np.log10(times)
    accs_percent = [a * 100 for a in accuracies]
    f1s = [baseline_results['f1'] * 1000,
           statistical_results['f1'] * 1000,
           lstm_results['f1'] * 1000]

    for i, (name, t, acc, f1, color) in enumerate(zip(model_names, times_log,
                                                        accs_percent, f1s, colors)):
        ax.scatter(t, acc, s=f1, alpha=0.6, color=color,
                  edgecolors='black', linewidths=2)
        ax.annotate(name, (t, acc), fontsize=11, fontweight='bold',
                   ha='center', va='center')

    ax.set_xlabel('Training Time (log scale)', fontsize=12, fontweight='bold')
    ax.set_ylabel('Accuracy (%)', fontsize=12, fontweight='bold')
    ax.set_title('Accuracy vs Speed Trade-off (Bubble size = F1-Score)',
                fontsize=14, fontweight='bold')
    ax.grid(alpha=0.3, linestyle='--')

    plt.tight_layout()
    plt.savefig('performance_tradeoff.png', dpi=300, bbox_inches='tight')
    plt.show()

    # Practical recommendations
    print("\n" + "="*70)
    print("PRACTICAL RECOMMENDATIONS")
    print("="*70)

    print("\nFor real-world deployment:")
    print("   1. Start with Naive Bayes for general use")
    print("   2. Monitor edge cases and difficult predictions")
    print("   3. Consider ensemble methods for improved accuracy")
    print("   4. Use confidence thresholds to flag uncertain predictions")
    print("   5. Apply deep learning selectively for challenging cases")

    print("\nIndustry-specific guidance:")
    print("   • Streaming platforms: statistical models for real-time processing")
    print("   • Batch processing: either approach is acceptable")
    print("   • Mobile applications: statistical models due to resource constraints")
    print("   • Research purposes: deep learning models for maximum performance")
    print("   • Explainable AI needs: statistical models are preferred")

    print("\nAnalysis complete. Visualizations have been saved.")
    print("="*70)

# Execute detailed analysis
detailed_analysis(baseline_results, statistical_results, lstm_results,
                  results_df, y_test)

print("\n" + "="*70)
print("PROJECT COMPLETE")
print("All models have been trained, evaluated, and analyzed.")
print("Refer to the generated PNG files for visualizations.")
print("="*70)
```

**Performance Analysis and Comparative Discussion**

***Quantitative Performance Comparison***

The results showed a surprising ranking of the three models. Baseline Logistic Regression had the highest accuracy at 87.40%, followed by the LSTM at 86.47% and Naive Bayes at 86.22%. The 1.18 percentage point difference shows that the simplest model performed better than the more complex ones. The baseline used only 3,000 word count features, which shows that selecting the right features can be more important than using a complex algorithm, especially for datasets with clear sentiment patterns. The LSTM, despite taking 111 times longer to train (1,311.7 seconds vs 11.8 seconds), had 0.93 percentage points lower accuracy, showing that the extra computation did not improve results.

Looking at precision, recall, and F1-scores gives more details. The baseline had balanced performance with 87.18% precision and 87.70% recall, showing it handled positive and negative reviews evenly. The LSTM had very high precision (90.46%) but low recall (81.54%), meaning it was too cautious in predicting positive reviews. Confusion matrix analysis confirms that the baseline made the most balanced errors, with 645 false positives and 615 false negatives, giving sensitivity of 87.70% and specificity of 87.10%. The LSTM, however, had 430 false positives and 923 false negatives, showing high specificity (91.40%) but low sensitivity (81.54%), which suggests it overfit to negative examples.

***Computational Efficiency Trade-offs***

The baseline is also much faster. Training took only 11.8 seconds compared to 1,311.7 seconds for the LSTM, a 111× difference, and it also achieved better accuracy. Running 50 iterations with the baseline takes about 10 minutes, while the LSTM requires over 18 hours. The LSTM’s high computational cost and lower performance make it impractical for this task.

***Model Selection Insights***

The baseline’s results highlight several key points. First, the IMDB dataset contains clear words like “excellent” and “terrible,” which simple word counts can capture without needing complex understanding of context. Second, 3,000 features seem to be the best choice. Expanding to 5,000 features in Naive Bayes, including TF-IDF and bigrams, added noise instead of improving performance. Third, the LSTM’s low recall shows it struggled to generalize and overfit to negative reviews.

Overall, baseline Logistic Regression is the best choice for IMDB sentiment analysis. It provides the highest accuracy, fastest training, balanced performance, and easy interpretation. The other models do not justify their extra complexity. This shows that strong baselines are important, and simpler models can perform better than complex ones when configured properly. The results also challenge the idea that deep learning is always superior and show the importance of matching model complexity to the task.
